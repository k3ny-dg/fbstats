```
#import modules/ tools

import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
import numpy as np
from collections import Counter
```

```
# open file and read all lines into a list variable

with open('superbowl.csv', 'r') as data_file:
  data_list = data_file.readlines()

#print(data_list)
```

```
# remove the 0th item (labels)

data_list.pop(0)

print(data_list)
```

```
# create empty lists

year_list = []
win_points_list = []
win_team_list = []
lose_points_list = []
mvp_list = []
lose_team_list = []
sb_num_list = []
```

```
# look at each item
# clean and split each item
# pick out items we care about
# append the relevant variables into lists

for item in data_list:
  date, sb_num, win_team, win_points, lose_team, lose_points, MVP, stadium, city, state = item.strip().split(',')

  year_list.append(date[-4:])
  win_points_list.append(int(win_points))
  win_team_list.append(win_team)
  lose_points_list.append(int(lose_points))
  lose_team_list.append(lose_team)
  mvp_list.append(MVP)
  sb_num_list.append(sb_num)

print(year_list)
print(win_points_list)
```

```
# add a graph paper grid to background
plt.figure(figsize= (44, 8))
sns.set_style('whitegrid')

# make a scatter plot with two lists and save it into a variable
my_graph = sns.lineplot(x = year_list, y = win_points_list)

# one line of code to add labels on the axis
my_graph.set(xlabel = 'Year (2020-1967)', ylabel = 'Final Points by Winning Team')

# display the plot
plt.show
```

**LINE PLOT EXPLANATION**

In this line plot we are looking at the data from all Super Bowls. The intent
of the graph was to see if the NFL is moving more towards an offensive approach, where there tend to be high scoring shoot outs, as opposed to a more balanced approach, which would yield more moderate scores. 

---


The variable on the horizontal axis is the year of each Super Bowl, from 1967 to 2020. 

The variable on the vertical axis is the winning (highest) score in each of those games. 

While the first couple decades of the Super Bowl tended to yield scores on the lower end, overall there does not appear to be any sort of trend in the data. In fact, in 2019, there was the lowest winning score on record in the Super Bowl in the game between the New England Patriots and the Los Angeles Rams (13-3). We can conclude from the data that it is unlikely that the NFL is moving to an offensive heavy approach. 

It looks like there may be an outlier, which was at the 1990 Super Bowl, which had a winning score of 55 points. We will confirm if this data point is an outlier below a Box Plot. 




```
# create a box plot to determine if there are outliers

plt.boxplot(win_points_list)
```

**BOX PLOT CONCLUSION**

It appears that there is a single outlier in this list of winning score, which is the highest winning score. Below, we will use the program to figure out which team had that highest score.

**(2) What team has won the most Super Bowls?**

```
# find the mode of win_team

print('The teams that have won the most Super Bowls are:')
print(' ')


def compute_mode(team_list):
    ''' finds the top value(s) in a list, for when there are multiple modes'''
    counts = {}
    maxcount = 0
    for team in team_list:
        if team not in counts:
            counts[team] = 0
        counts[team] += 1
        if counts[team] > maxcount:
            maxcount = counts[team]

    for team, count in counts.items():
        if count == maxcount:
            print(team, 'with a total of', count)

compute_mode(win_team_list)


```

**(3) What team(s) have lost the most Super Bowls?**

```
# reuse the compute_mode function created above to find this answer

print('The teams that have lost the most Super Bowls are:')
print(' ')

compute_mode(lose_team_list) 
```

**(4) Who had the highest winning score? What was it?**

```
# calcuate what the max win_points is and find out what team had that score

high_score= max(win_points_list)

index = win_points_list.index(high_score)

print('The team with the highest winning score was the', win_team_list[index], 'with', high_score, 'points.')
```

**(5) What is the average point differential in a game?**


```
# avg (win_points - lose_points) 

difference_list = []

for index in range(len(win_points_list)):
  diff = win_points_list[index] - lose_points_list[index]
  difference_list.append(diff)

avg_diff = stats.mean(difference_list)

print(f'The average point differential is {avg_diff:.1f} points.')
```

**(6) Which were close games (within one touchdown)?**



```
# find scores in difference_list < 7
# use their index position to print the SB # and the teams that played
# structured as: 
# print(sb_num, winning team, 'vs.', losing team)

print('The Super Bowls that were close games:')
print(' ')
for index in range(len(difference_list)):
  if difference_list[index] < 7:
    print(sb_num_list[index], win_team_list[index], 'vs.', lose_team_list[index])
```

**(7) Which player has won MVP the most times?**

```
# find the mode of mvp_list

top_mvp = stats.mode(mvp_list)

print('The player that has won the most MVP awards is:', top_mvp)
```
